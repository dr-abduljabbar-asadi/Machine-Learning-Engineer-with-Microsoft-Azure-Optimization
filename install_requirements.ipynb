# Install Project Requirements

Run this notebook **once** before running `udacity-project.ipynb`, using the **same kernel** you will run the project with.

It installs the Azure ML SDK v1 training packages (HyperDrive + AutoML), the widgets, and the ML libraries the project needs. On the Udacity lab (Python 3.10) these are not all present by default, which causes errors like `No module named 'azureml.train'` or `No module named 'joblib'`.

**After it finishes, RESTART THE KERNEL** (Kernel > Restart), then open and run `udacity-project.ipynb`.

```python
# Show the installed azureml-core version. The training packages are pinned to
# this exact version to avoid dependency conflicts.
import azureml.core

core_version = azureml.core.VERSION
print("azureml-core version:", core_version)
```

```python
# Install the Azure ML SDK v1 training packages, pinned to the azureml-core version.
%pip install --quiet \
    "azureml-train-core=={core_version}" \
    "azureml-train-automl-client=={core_version}" \
    "azureml-train-automl-runtime=={core_version}" \
    "azureml-widgets=={core_version}"
```

```python
# Install the general ML libraries used by the notebook and train.py.
%pip install --quiet joblib scikit-learn pandas numpy
```

```python
print("Install finished.")
print("Now RESTART THE KERNEL (Kernel > Restart), then run udacity-project.ipynb from the top.")
```
