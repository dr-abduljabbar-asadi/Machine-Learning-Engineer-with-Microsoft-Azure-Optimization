# Preflight Check

Run this notebook **before** `udacity-project.ipynb` to confirm your Azure ML lab
environment can run the project. Each cell prints a clear **PASS** or **FAIL**.

> Make sure the selected kernel is **Python 3.8 - AzureML** (kernel picker, top-right).

```python
# 1) Python version must be 3.8 (Azure AutoML does not support 3.10+).
import sys

print("Python:", sys.version)
if sys.version_info[:2] == (3, 8):
    print("PASS: running on Python 3.8")
else:
    print("FAIL: wrong kernel. Select 'Python 3.8 - AzureML' and re-run.")
```

```python
# 2) Azure ML SDK and AutoML must be importable in this kernel.
try:
    import azureml.core
    import azureml.train.automl  # noqa: F401
    print("azureml.core version:", azureml.core.VERSION)
    print("PASS: azureml.core and azureml.train.automl import OK")
except Exception as e:
    print("FAIL: could not import the AutoML SDK ->", repr(e))
```

```python
# 3) Workspace loads and the curated sklearn environment exists.
try:
    from azureml.core import Workspace, Environment

    ws = Workspace.from_config()
    print("Workspace:", ws.name)

    env_name = "AzureML-sklearn-0.24-ubuntu18.04-py37-cpu"
    names = list(Environment.list(workspace=ws).keys())
    sklearn_envs = [n for n in names if "sklearn" in n.lower()]

    if env_name in names:
        print(f"PASS: curated env '{env_name}' is available")
    else:
        print(f"FAIL: '{env_name}' not found. Available sklearn envs:")
        for n in sklearn_envs:
            print("  -", n)
except Exception as e:
    print("FAIL: workspace/environment check ->", repr(e))
```

```python
# 4) The public dataset is reachable and clean_data works.
try:
    import pandas as pd
    from train import clean_data

    data_url = (
        "https://automlsamplenotebookdata.blob.core.windows.net/"
        "automl-sample-notebook-data/bankmarketing_train.csv"
    )
    df = pd.read_csv(data_url)
    x, y = clean_data(df)
    print("Rows:", len(df), "| Features after clean_data:", x.shape[1])
    print("PASS: data read and clean_data OK")
except Exception as e:
    print("FAIL: data/clean_data check ->", repr(e))
```
